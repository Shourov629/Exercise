# Scatter Plot

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.arange(0,10)
y = np.arange(11,21)

plt.scatter(x,y)
plt.xlabel("X axis")
plt.ylabel("Y axis")
plt.title("Scatter Plot")
print(x,y)
plt.show()
```

# Bar Plot

```python
c = ["r","g","b","y","m","r","g","b","y","m"]
plt.bar(x,y,color=c)
```

```python
y = x**2
y
```

# Plot

```python
plt.plot(x,y,'b*', linestyle='solid', linewidth=.1, markersize=12)
```

```python
plt.subplot(2,2,1)
plt.plot(x,y,'b*', linestyle='solid', linewidth=.1, markersize=5)
plt.subplot(2,2,2)
plt.plot(x,y,'r*-', linewidth=.1, markersize=5)
plt.subplot(2,2,3)
plt.plot(x,y,'yo--',markersize=5)
plt.subplot(2,2,4)
plt.plot(x,y,'go-',linewidth=.1, markersize=5)
```

# Generation of Sine wave

```python
x = np.arange(0, 8*np.pi, 0.1)
cur = np.sin(x)
plt.plot(cur, 'go-', linewidth=0.15, markersize=3)
plt.show()
```

```python
x = np.arange(0, 0.75*np.pi, 0.1)
cur = np.sin(x)
plt.bar(x,cur,color=['r','g'])
plt.show()
```

# Histogram

```python
xy = [3,2,40,45,51,53,33,35,39,65,8,23,29,89,82,87]
plt.hist(xy)
```

# Pi Chart

```python
labels = 'Python', 'Ruby', 'C++', 'Java'
sizes = [245, 130, 325, 171]
colors = ['gold', 'yellowgreen', 'lightcoral', 'lightskyblue']
explode = (0, 0, 0, 0)

plt.pie(sizes,explode=explode,labels=labels,colors=colors,autopct='%1.1f%%',shadow=True)
plt.axis('equal')
plt.show()
```

# Box Plot

```python
data = [np.random.normal(0,std,100) for std in range(1,4)]

plt.boxplot(data,vert=True,patch_artist=False)
plt.show()
```
