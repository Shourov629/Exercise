# Use of dropna

```python
import pandas as pd
csv_df = pd.read_csv("D:\\ML & DS\\Exercise\\Pandas\\missing_value_of_wine_dataset.csv")
csv_df
```

```python
#removing all the NaN values
csv_df.dropna()
```

```python
#removing NaN values by row (axis = 1 for col)
csv_df.dropna(axis=0)
```

```python
#removing NaN value one by one
var = csv_df.dropna(how='any')
var
```

```python
#removing a entire row which contains only NaN
var = csv_df.dropna(how='all')
var
```

```python
#removing NaN from a particular column
var = csv_df.dropna(subset="density")
var
```

```python
#threshold value
var = csv_df.dropna(thresh=1)
var
```

# Use of fillna

```python
#fill all the NaN by something
var = csv_df.fillna('any')
var
```

```python
#fill NaN from selective columns with selective values
var = csv_df.fillna({"density":"python","residual_sugar":"Hello","citric_acid":"World"})
var
```

```python
#filling all the NaN by forward value using method = ffill (by backward value using "bfill")
var = csv_df.fillna(method="ffill")
var
```

# Use "replace()"

```python
csv_df.replace(to_replace="red", value="hellooo")
```

```python
csv_df.replace([1,2,3,4,5,6,7,8,9], "Serial")
```
