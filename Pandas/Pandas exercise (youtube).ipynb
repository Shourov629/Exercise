# Series

```python
import pandas as pd

x = [1,3,3,4,5,67,8,34,5]
y = [14,35,67,84,95,67,8,34,5]

var_1 = pd.Series(x,index=[1,2,3,4,5,6,7,8,9], dtype='float')
var_2 = pd.Series(y)
print(var_1)
print(var_2)
```

```python
var = pd.Series(12, index=[1,2,3,4,5])
var1 = pd.Series(31, index=[1,2,3,4,5,6,7,8])
print(var+var1)
```

```python
d = {"Name":['python', 'C', 'C++', 'Java'], "Por":[12, 21, 32, 23], "Rank":[1, 4, 3, 2]}
dick = pd.Series(d)

print(dick)
```

# Arithmetic Operations

```python
nli = {"F":[14,35,67,84,95,67,8,34,5], "S":[1,3,3,4,5,67,8,34,5]}

nd = pd.DataFrame(nli)

nd["SUM"] = nd["F"] + nd["S"]
nd["SUB"] = nd["F"] - nd["S"]
nd["MUL"] = nd["F"] * nd["S"]
nd["DIV"] = nd["F"] / nd["S"]

nd
```

```python
nd["DIV < 20"] = nd["DIV"] <= 20

nd
```

```python
nd["MUL > 99"] = nd["MUL"] >= 99

nd
```

# Insertion

```python
nd.insert(1,"F<50",nd["F"]<50)
nd
```

```python
nd.insert(3, "Slicing", nd["SUM"][0:6])
nd
```

```python
nd.insert(4, "Slicing_2", nd["SUM"][0:6])
nd
```

# Deletion

```python
after_delete = nd.pop("Slicing_2")
after_delete
```
